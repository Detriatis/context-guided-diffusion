```python
sims = np.load("data/datasets/COMPAS_tanimoto_count_sims.npy")
```

```python
sims_11 = sims[np.ix_(data["n_rings"] == 11, data["n_rings"] == 11)]

clusters = pd.Series(
    SpectralClustering(
        n_clusters=2,
        n_components=5,
        affinity='precomputed',
        random_state=1,
        assign_labels='kmeans',
        n_jobs=-1,
    ).fit_predict(sims_11)
)

display(clusters.value_counts(normalize=True).sort_index())

# additionally split off validation set

train_index = clusters.to_numpy().astype(bool)

val_clusters = pd.Series(
    SpectralClustering(
        n_clusters=2,
        affinity='precomputed',
        random_state=1,
        assign_labels='kmeans',
        n_jobs=-1,
    ).fit_predict(sims_11[np.ix_(~train_index, ~train_index)])
)

display(val_clusters.value_counts(normalize=True).sort_index())

# use largest cluster as training, second largest as test set and third largest as validation set

val_clusters[val_clusters == 0] = 2

clust = clusters.to_numpy()
clust[train_index] = 0
clust[~train_index] = val_clusters

data["cluster_split"] = -1
data.loc[data["n_rings"] == 11, "cluster_split"] = clust
data["cluster_split"].value_counts().sort_index()
```

```python
# get random split to match cluster split

random_list = list(range(len(data)))
random.shuffle(random_list)

n_train = len(data[data["cluster_split"]==0])
n_val = len(data[data["cluster_split"]==1])
n_test = len(data[data["cluster_split"]==2])

random_split_idx = np.zeros(len(data))
random_split_idx[random_list[:n_train]] = 0 
random_split_idx[random_list[n_train:n_train+n_val]] = 1
random_split_idx[random_list[n_train+n_val:n_train+n_val+n_test]] = 2
random_split_idx[random_list[n_train+n_val+n_test:]] = -1

data["random_split"] = random_split_idx
data["random_split"].value_counts().sort_index()
```

```python
mapper = umap.UMAP(n_neighbors=1000, min_dist=0.5, metric="precomputed").fit(1-sims)
umap.plot.points(mapper, labels=data["cluster_split"], alpha=0.4, theme="fire")
umap.plot.points(mapper, labels=data["random_split"], alpha=0.4, theme="fire")
```

```python
def kernel_2samp_test(train_idx, test_idx, sim_mat):
    """
    Implements the unbiased estimator of the maximum mean discrepancy statistic for a training index splitting a given kenrel similarity matrix.
    """
    train_mean = (sim_mat[np.ix_(train_idx, train_idx)].sum() - train_idx.sum()) / (train_idx.sum() * (train_idx.sum() - 1))
    test_mean = (sim_mat[np.ix_(test_idx, test_idx)].sum() - test_idx.sum()) / (test_idx.sum() * (test_idx.sum() - 1))
    train_test_mean = sim_mat[np.ix_(train_idx, test_idx)].mean()
        
    mmd_squared = train_mean + test_mean - 2 * train_test_mean
    return np.sqrt(mmd_squared)

def get_cov_stats(split_col):
    distances = np.zeros((4, 4))

    for i in range(-1, 3):
        for j in range(i+1, 3):

            i_idx = (data[split_col] == i).values
            j_idx = (data[split_col] == j).values
            
            #sims_reduced = sims[np.ix_(comb_index, comb_index)]
            #comb_index = np.logical_or(i_idx, j_idx)
            #sims_reduced = sims[np.ix_(comb_index, comb_index)]

            cov_shift = kernel_2samp_test(i_idx, j_idx, sims)

            distances[i, j] = cov_shift
            distances[j, i] = cov_shift

    sns.heatmap(distances, annot=True, square=True)
    plt.title(f"Covariate Shift Between Training (0), Validation (1) and Test (2) Sets for {split_col}\n")
```

```python
def get_label_stats(split_col):
    label_stats = dict()

    for target_col in target_cols:

        ks_stats = np.zeros((4, 4))
        for i in range(-1, 3):
            for j in range(i+1, 3):

                i_ys = data.loc[data[split_col] == i, target_col].to_numpy()
                j_ys = data.loc[data[split_col] == j, target_col].to_numpy()

                ks_stat = ks_2samp(i_ys, j_ys)[0]
                ks_stats[i, j] = ks_stat
                ks_stats[j, i] = ks_stat

        label_stats[target_col] = ks_stats
        
    fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(10, 10))

    for i , target_col in enumerate(target_cols):
        sns.heatmap(label_stats[target_col], annot=True, square=True, ax=axs[i//3, i%3], cbar=False)
        axs[i//3, i%3].title.set_text(target_col + "\n")
    fig.delaxes(axs[1,2])

    plt.tight_layout()

    return label_stats
```
